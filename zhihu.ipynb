```python
def get_response(url):
    """
    获取对应url内容
    """
    # 设置请求头，模拟请求者是浏览器，防止被拒绝访问
    headers = { 
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36',
    }
    # 这里捕捉异常是防止请求过程中出错，导致程序崩溃。
    try:
        response = requests.get(url=url, headers=headers)
    except Exception as e:
        return None
    # 如果请求成功后放回的状态码是200就返回对应的数据，否则返回None
    return response.text if response.status_code is 200 else None


def parse_response(response):
    """
    由于返回的数据是Json格式的，所以需要相应的解析，
    这里这获取作者名和回答的内容
    """
    json_datas = json.loads(response)['data']
    comments = []
    
    for item in json_datas:
        try:
            comment = []
            comment.append(item['author']['name'])
            comment.append(item['content'])
            comments.append(comment)
        except Exception as e:
            pass
    return comments


def save_data(comments):
    """
    保存数据
    """
    filename = 'comments2.csv'
    df = pd.DataFrame(comments, columns=['name', 'content'])
    df.to_csv(filename, mode='a', index=False, sep=',')
    
    
def check_is_end(response):
    """
    判断是否结束爬取
    """
    return json.loads(response)['paging']['is_end']

def get_next_url(response):
    """
    获取下一个url
    """
    return json.loads(response)['paging']['next']
    
    
# 问题id，理论上只需要改变这里的id就可以爬取对应的回答
question_id = '28912859'
# 拼接url，注意这里的后面的offset=0表示从最初开始爬取
url = 'https://www.zhihu.com/api/v4/questions/'+ question_id + '/answers?include=data%5B%2A%5D.is_normal%2Cadmin_closed_comment%2Creward_info%2Cis_collapsed%2Cannotation_action%2Cannotation_detail%2Ccollapse_reason%2Cis_sticky%2Ccollapsed_by%2Csuggest_edit%2Ccomment_count%2Ccan_comment%2Ccontent%2Ceditable_content%2Cvoteup_count%2Creshipment_settings%2Ccomment_permission%2Ccreated_time%2Cupdated_time%2Creview_info%2Crelevant_info%2Cquestion%2Cexcerpt%2Crelationship.is_authorized%2Cis_author%2Cvoting%2Cis_thanked%2Cis_nothelp%2Cis_labeled%2Cis_recognized%2Cpaid_info%3Bdata%5B%2A%5D.mark_infos%5B%2A%5D.url%3Bdata%5B%2A%5D.author.follower_count%2Cbadge%5B%2A%5D.topics&limit=5&offset=0&platform=desktop&sort_by=default'

response = get_response(url)

if response is not None:
    contents = []
    while check_is_end(response) != True:
        datas = parse_response(response)
        contents.append(datas)
        url = get_next_url(response)
        response = get_response(url)
    print(contents[:8])
```
